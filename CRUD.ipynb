# Projeto de estudos CRUD (Create, Read, Update, Delete) de SQL com o SQLite3

```python
# Importando o pacote sqlite3
import sqlite3
```

```python
class CRUD:
  def __init__ (self):
    self.name = 'data.db'
    self.conn = sqlite3.connect(self.name)
    self.cur = self.conn.cursor()

    # Chamando o criador da tabela user
    self.start()


  # Criador da tabela user [id, user, email e password]
  def start(self):
    sql = '''CREATE TABLE IF NOT EXISTS user (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    user VARCHAR(64) NOT NULL,
    email VARCHAR(64) NOT NULL,
    password VARCHAR(128) NOT NULL
    )'''

    self.cur.execute(sql)
    self.conn.commit()


  # Método para inserir dados
  def create(self, user, email, password):

    sql = f''' SELECT id FROM user WHERE email = \'{email}\' OR user = \'{user}\' '''
    self.cur.execute(sql)

    data = self.cur.fetchall()

    if len(data) != 0:
      raise Exception(f"The user {user} and email {email} already exists")
      return False

    sql = f''' INSERT INTO user (user, email, password) VALUES (\'{user}\', \'{email}\', \'{password}\')'''

    self.cur.execute(sql)
    self.conn.commit()


  # Método para ler o id e user a partir do email e password
  def read(self, email, password):
    sql = f''' SELECT id, user FROM user WHERE email = \'{email}\' AND password = \'{password}\' '''

    self.cur.execute(sql)

    data = self.cur.fetchall()

    return data


  # Método para atualizar o user a partir do email e password
  def update(self, user, email, password):

    sql = f''' SELECT id FROM user WHERE email = \'{email}\' AND password = \'{password}\' '''

    self.cur.execute(sql)

    data = self.cur.fetchall()

    if len(data) == 0:
      raise Exception(f"The email {email} does not exists or password is wrong")
      return False

    sql = f''' UPDATE user SET user = \'{user}\' WHERE email = \'{email}\' AND password = \'{password}\' '''

    self.cur.execute(sql)
    self.conn.commit()


  # Método para deletar o usuário a partir do email e password
  def delete(self, email, password):
    sql = f''' SELECT id FROM user WHERE email = \'{email}\' AND password = \'{password}\' '''

    self.cur.execute(sql)

    data = self.cur.fetchall()

    if len(data) == 0:
      raise Exception(f"The email {email} does not exists or password is wrong")
      return False

    sql = f''' DELETE FROM user WHERE email = \'{email}\' AND password = \'{password}\' '''

    self.cur.execute(sql)
    self.conn.commit()


  # Sistema de debug para ler todo os registros da base de dados
  def READ_DEBUG(self):
    sql = f''' SELECT * FROM user '''

    self.cur.execute(sql)

    return self.cur.fetchall()
```

```python
# Instanciando o CRUD
crud = CRUD()
```

```python
# Inserindo dados
try:
  crud.create('user_01', 'example1@gmail.com', '123456')
  crud.create('user_02', 'example2@gmail.com', '123456')
except Exception as e:
  print(e)
```

```python
# Obtendo o erro
try:
  crud.create('user_01', 'example1@gmail.com', '123456')
  crud.create('user_02', 'example2@gmail.com', '123456')
except Exception as e:
  print(e)
```

```python
# Obtendo todo os registros
crud.READ_DEBUG()
```

```python
# Lendo apenas um registro
crud.read('example1@gmail.com', '123456')
```

```python
# Atualizando o registro
crud.update('__myuser__', 'example1@gmail.com', '123456')
```

```python
# Obtendo um erro ao atualizar o registro
try:
  crud.update('__myuser__', 'example3@gmail.com', '123456')
except Exception as e:
  print(e)
```

```python
crud.READ_DEBUG()
```

```python
# Apagando um registro
try:
  crud.delete('example1@gmail.com', '123456')
except Exception as e:
  print(e)
```
